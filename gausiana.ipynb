```python
import matplotlib.pyplot as plt
import numpy as np
import math

# custom function
def sigmoid(x):
    a = 1
    c = 0.25
    s0 = 1
    return a * math.exp( -(x-s0) ** 2 / 2 * c ** 2 )

# define vectorized sigmoid
sigmoid_v = np.vectorize(sigmoid)

# test
x_axis = np.arange(0,10, 0.001)
y_axis = np.arange(0,10, 0.001)
result = sigmoid_v(y_axis)
plt.plot(x_axis, result)
plt.show()


```
